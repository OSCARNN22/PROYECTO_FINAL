# Visualizaciones con Plotly

**Gráfico 1 - Margen de Ganancia por Año**



El gráfico muestra la evolución del margen de ganancia promedio por año.

```python
yearly_profit_margin = (
    df.groupby("order_year")["profit_margin"]
    .mean()
    .round(2)
    .reset_index()
)

fig = px.line(
    yearly_profit_margin,
    x="order_year",
    y="profit_margin",
    title="Profit Margin by Year",
    markers=True,
    text="profit_margin"
)

fig.update_traces(
    line_color="red",
    texttemplate="%{text:.2f}%",
    textposition="top center"
)

fig.update_layout(
    xaxis_title="Year",
    yaxis_title="Profit Margin (%)",
    template="plotly_white"
)

fig.show()
```

**Gráfico 2 - Ventas por Categoría**

Este gráfico interpreta que la categoría Technology es la que registra el mayor volumen de ventas, seguida por Furniture y Office Supplies. Esto indica que los productos tecnológicos representan la principal fuente de ingresos dentro del conjunto de datos analizado.

```python
category_sales = (
    df.groupby("category")["sales"]
    .sum()
    .sort_values(ascending=False)
    .reset_index()
)

fig = px.bar(
    category_sales,
    x="category",
    y="sales",
    title="Sales by Category",
    color="category",
    text_auto=".2s"
)

fig.update_layout(
    xaxis_title="Category",
    yaxis_title="Total Sales",
    template="plotly_white"
)

fig.show()
```

**Gráfico 3 - Tendencia de ventas Mensuales**

Este gráfico representa la tendencia mensual de las ventas a lo largo del período analizado, permitiendo:
- Identificar meses con mayores o menores ventas.
- Detectar tendencias de crecimiento o disminución.
- Observar posibles patrones estacionales en el comportamiento de las ventas.

```python
monthly_sales_chart["month_label"] = (
    monthly_sales_chart["order_month"].str[:3]
    + "-"
    + monthly_sales_chart["order_year"].astype(str)
)

fig = px.line(
    monthly_sales_chart,
    x="month_label",
    y="sales",
    title="Monthly Sales Trend",
    markers=True
)

fig.update_layout(
    xaxis_title="Month",
    yaxis_title="Sales",
    xaxis_tickangle=-45,
    template="plotly_white"
)

fig.show()
```

**Gráfico 4 - Los 5 principales estados con más ventas**

Este gráfico permite identificar:

- Los estados con mayor contribución al volumen de ventas.
- La concentración geográfica del negocio.
- Regiones con mayor importancia comercial.

```python
top_states = (
    df.groupby("state")["sales"]
    .sum()
    .sort_values(ascending=False)
    .head(5)
    .reset_index()
)

fig = px.bar(
    top_states,
    x="state",
    y="sales",
    title="Top 5 States by Sales",
    color="sales",
    text="sales"
)

fig.show()
```

**Gráfico 5 - Distribución de segmentos de clientes**

Este gráfico permite identificar qué tipo de clientes tiene mayor participación en los ingresos del negocio.

```python
segment_sales = (
    df.groupby("segment")["sales"]
    .sum()
    .reset_index()
)

fig = px.pie(
    segment_sales,
    values="sales",
    names="segment",
    title="Sales Distribution by Customer Segment",
    color_discrete_sequence=["yellow", "red", "blue"],
    hole=0.
)

fig.update_traces(
    textinfo="percent+label",
    textposition="inside"
)

fig.show()
```

# Consultas SQL con DuckDB

Importar la librería DuckDB

```python
import duckdb
```

Crear una base de datos en memoria

```python
con = duckdb.connect(database=":memory")
```

Registra el DataFrame df como una tabla virtual llamada store.

A partir de ese momento puedes escribir consultas SQL como si store fuera una tabla de una base de datos.

```python
con.register("store",df)
```

Se puede dividir en dos partes:

- con.execute(): ejecuta la consulta SQL sobre la tabla store.
- fetch_df(): devuelve el resultado como un DataFrame de Pandas.

```python
con.execute("SELECT * FROM store").fetch_df()
```

**1. Inspección de datos (SELECT + LIMIT)**

Permite visualizar una muestra de los 10 registros del dataset.

  ***Objetivo:**
Inspeccionar la estructura del dataset y verificar que los datos se hayan cargado correctamente.*

```python
con.execute("""
SELECT *
FROM store
LIMIT 10
""").fetch_df()
```

**2. Filtrar información (WHERE)**

Mostrar las ventas con una ganancia superior a 3,000.

  ***Objetivo:**
  Identificar las ventas con mayor rentabilidad.*

```python
con.execute("""
SELECT
    "Order_ID",
    "Customer_Name",
    Category,
    Sales,
    Profit
FROM store
WHERE Profit > 3000
""").fetch_df()
```

**3. Agrupación y agregación (GROUP BY)**

Calcular las ventas totales por categoría.

  ***Objetivo:**
      Comparar el volumen de ventas entre las diferentes categorías de productos.*

```python
con.execute("""
SELECT
    Category,
    SUM(Sales) AS Total_Sales
FROM store
GROUP BY Category
""").fetch_df()
```

**4. Ranking (ORDER BY)**

Obtener los cinco estados con mayores ventas.

  ***Objetivo:**
Identificar los estados con mayor volumen de ventas.*

```python
con.execute("""
SELECT
    State,
    SUM(Sales) AS Total_Sales
FROM df
GROUP BY State
ORDER BY Total_Sales DESC
LIMIT 5;
""").fetch_df()
```

**5. Consulta combinada (WHERE + GROUP BY + ORDER BY + cálculo)**

Calcular las ventas y ganancias por categoría considerando únicamente las ventas superiores a 100 dólares.

  ***Objetivo:**
Analizar la rentabilidad de cada categoría considerando únicamente ventas de mayor valor.*

```python
con.execute("""
SELECT
    Category,
    SUM(Sales) AS Total_Sales,
    SUM(Profit) AS Total_Profit,
    ROUND(AVG(Profit),2) AS Average_Profit
FROM df
WHERE Sales > 100
GROUP BY Category
ORDER BY Total_Profit DESC;
""").fetch_df()
```
